```python
data = pd.read_csv("Customers.csv")
data.head()
```

```python
# Below code corrected two rows  date_data in one format
# Column name - birthdate, acct_open_date
# format - 8/26/1961 = MM/dd/yyyy , 07-03-1915 = dd-mm-yyyy
# output data dd-mm-yyyy
```

```python
import pandas as pd
from datetime import datetime

df = pd.DataFrame(data)

# Function to parse and convert dates
def convert_date_format(date_str):
    for fmt in ("%m/%d/%Y", "%d-%m-%Y"):  # Possible date formats
        try:
            return datetime.strptime(date_str, fmt).strftime("%d-%m-%Y")
        except ValueError:
            continue
    return None  # Return None if no format matches

# Update the birthdate column with the formatted values
df["birthdate"] = df["birthdate"].map(convert_date_format)
df["acct_open_date"] = df["acct_open_date"].map(convert_date_format)

# Verify the results
print(df.head())
```
